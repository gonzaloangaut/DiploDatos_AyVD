```python
## Tabla de contingencia
# muestra la relación entre "df.work_province" y "df.profile_studies_level"
pd.crosstab(df.work_province, df.profile_studies_level)#, normalize=True) #no están ordenados
pd.crosstab(df.work_province, df.profile_studies_level)[sorted_studies_levels]#ordenados
```

```python
# veamos la tabla de contingencia en un mapa de calor
fig = plt.figure(figsize=(8,6))
exp = pd.crosstab(df.work_province,
                  df.profile_studies_level)[sorted_studies_levels]
seaborn.heatmap(exp, annot=True, fmt='g') # mapa de calor
plt.xticks(rotation=30)
plt.yticks(rotation=0)
seaborn.despine()
```

```python
## Tabla de contingencia normalizada, la suma de los valores da 1, son proporciones, estima probabilidad
pd.crosstab(df.profile_age, df.profile_studies_level, normalize=True)[sorted_studies_levels]
```

## Varias numéricas

```python
max_sal = 10000000
# grafica relaciones entre variables
seaborn.pairplot(data=df[df.salary_monthly_NETO <= max_sal],
                 y_vars=['profile_age', 'salary_monthly_BRUTO'], # variables en el eje y
                 x_vars=['salary_monthly_NETO', 'salary_monthly_BRUTO'], # variables en el eje x
                 aspect=2, height=3)
```

```python
fig = plt.figure(figsize=(8,4))
# visualizar densidades en scatter plots con muchos puntos superpuestos
# LA FORMA DE HACERLO EN LA OTRA CELDA ES MEJOR
seaborn.jointplot(data=df[df.salary_monthly_NETO <= max_sal], x='salary_monthly_NETO',
                  y='profile_age', kind='hex', height=6)
plt.axvline(df[df.salary_monthly_NETO <= max_sal]['salary_monthly_NETO'].mean(), color='orangered',
            linestyle='--', label='Media Salario')
plt.axhline(df[df.salary_monthly_NETO <= max_sal]['profile_age'].mean(), color='blue',
            linestyle='--', label='Media Edad')
plt.legend()
plt.colorbar()
# Remove scietific notation
plt.ticklabel_format(style='plain', axis='y')
```

```python
# Filtro
df_filtrado = df[df.salary_monthly_NETO <= max_sal]

# Crear el jointplot
g = seaborn.jointplot(
    data=df_filtrado,
    x='salary_monthly_NETO',
    y='profile_age',
    kind='hex',
    height=6
)

# Agregar líneas de la media al eje principal
g.ax_joint.axvline(df_filtrado['salary_monthly_NETO'].mean(), color='orangered', linestyle='--', label='Media Salario')
g.ax_joint.axhline(df_filtrado['profile_age'].mean(), color='blue', linestyle='--', label='Media Edad')

# Agregar leyenda
g.ax_joint.legend()

# Quitar notación científica
g.ax_joint.ticklabel_format(style='plain', axis='x')
g.ax_joint.ticklabel_format(style='plain', axis='y')

# Mostrar el gráfico
plt.show()
```

```python
# densidad bivariada
# Crear el jointplot tipo KDE
g = seaborn.jointplot(
    data=df[df.salary_monthly_NETO <= max_sal],
    x='salary_monthly_NETO',
    y='profile_age',
    kind='kde',
    height=6
)

# Quitar notación científica en ambos ejes
g.ax_joint.ticklabel_format(style='plain', axis='x')
g.ax_joint.ticklabel_format(style='plain', axis='y')

plt.show()
```
